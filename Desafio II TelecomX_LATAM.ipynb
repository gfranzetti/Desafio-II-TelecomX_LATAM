```python
# Antes que nada SUBIR EL ARCHIVO "TelecomX_Data.json" al ambiente de trabajo
url = '/content/sample_data/TelecomX_Data.json'
```

```python
datos.info()
```

#🔧 Transformación

```python
datos_nor_base = datos.iloc[:, [0,1]]
datos_nor_base
```

```python
datos_nor_base.info()
```

```python
datos_nor_customer = pd.json_normalize(datos['customer'])
datos_nor_customer
```

```python
datos_nor_customer.info()
```

```python
datos_nor_phone = pd.json_normalize(datos['phone'])
datos_nor_phone
```

```python
datos_nor_phone.info()
```

```python
datos_nor_internet = pd.json_normalize(datos['internet'])
datos_nor_internet
```

```python
datos_nor_internet.info()
```

```python
datos_nor_account = pd.json_normalize(datos['account'])
datos_nor_account
```

```python
datos_nor_account.info()
```

```python
# Calcular el total de consumo

# Reemplaza espacios con "0"
datos_nor_account['Charges.Total'] = datos_nor_account['Charges.Total'].replace(' ', '0')

# Poner '0' donde hay NAN
datos_nor_account['Charges.Total'] = datos_nor_account['Charges.Total'].fillna(0)

# Converti 'object' a 'Float64'
datos_nor_account['Charges.Total'] = datos_nor_account['Charges.Total'].astype("float64")
# Chequear la converison

datos_nor_account.info()

# Chequear que no queden NAN

sum(pd.isna(datos_nor_account['Charges.Total']))
```

```python
full_datos = pd.concat([datos_nor_base,datos_nor_customer,datos_nor_phone,datos_nor_internet,datos_nor_account], axis=1)
full_datos
```

```python
full_datos.info()
```

```python
full_datos.to_csv('/content/sample_data/full_datos.csv' , index=False)
```

```python
# Diccionario de datos
#
#- `customerID`: número de identificación único de cada cliente
#- `Churn`: si el cliente dejó o no la empresa
#- `gender`: género (masculino y femenino)
#- `SeniorCitizen`: información sobre si un cliente tiene o no una edad igual o mayor a 65 años
#- `Partner`: si el cliente tiene o no una pareja
#- `Dependents`: si el cliente tiene o no dependientes
#- `tenure`: meses de contrato del cliente
#- `PhoneService`: suscripción al servicio telefónico
#- `MultipleLines`: suscripción a más de una línea telefónica
#- `InternetService`: suscripción a un proveedor de internet
#- `OnlineSecurity`: suscripción adicional de seguridad en línea
#- `OnlineBackup`: suscripción adicional de respaldo en línea
#- `DeviceProtection`: suscripción adicional de protección del dispositivo
#- `TechSupport`: suscripción adicional de soporte técnico, menor tiempo de espera
#- `StreamingTV`: suscripción de televisión por cable
#- `StreamingMovies`: suscripción de streaming de películas
#- `Contract`: tipo de contrato
#- `PaperlessBilling`: si el cliente prefiere recibir la factura en línea
#- `PaymentMethod`: forma de pago
#- `Charges.Monthly`: total de todos los servicios del cliente por mes
#- `Charges.Total`: total gastado por el cliente
#
```

```python
# Comparar edad de quienes abandonaron segun edad (menor de 65 o mayor de 65)

suma_por_SeniorCitizen = clientes_sin_faltante.groupby('Churn')['SeniorCitizen'].sum()
suma_por_SeniorCitizen

plt.ylabel('Cantidad de clientes')
plt.title('Edad de los Clientes')
categorias = ['Menores', 'Mayores']
plt.xlabel('Edad Clientes')
plt.bar(categorias, suma_por_SeniorCitizen, color='skyblue')
plt.show()
```

```python
# Inicio del servicio de SQL para querys

import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
engine = create_engine('sqlite:///:memory:')

clientes_sin_faltante.to_sql('clientes', engine, index=False)
```

```python
# Clientes que abandonaron segun el tipo de servicio que tenian contratado

servicios = clientes_sin_faltante['InternetService'].unique().astype(str)

clientes_servicios = []

for x in servicios:
  query = 'SELECT * FROM clientes WHERE Churn = "Yes" AND InternetService = '+'"'+x+'"'
  abandonaron = pd.read_sql(sql=text(query), con=engine.connect())
  clientes_servicios.append(len(abandonaron))

plt.ylabel('Cantidad de clientes que abandonaron')
plt.title('Clientes que abandonaron segun el tipo de servicio')
categorias = ['DSL', 'Fibra Optica', 'Ninguno']
plt.xlabel('Tipo de Servicio')
plt.bar(categorias, clientes_servicios, color='skyblue')
plt.show()
```

```python
# Clientes que abandonaron segun el tipo de soporte que tenian contratado

servicios = clientes_sin_faltante['TechSupport'].unique().astype(str)

clientes_TechSupport = []

for x in servicios:
  query = 'SELECT * FROM clientes WHERE Churn = "Yes" AND TechSupport = '+'"'+x+'"'
  abandonaron = pd.read_sql(sql=text(query), con=engine.connect())
  clientes_TechSupport.append(len(abandonaron))

plt.ylabel('Cantidad de clientes que abandonaron')
plt.title('Clientes que abandonaron segun el tipo de Soporte')
categorias = ['Si', 'No', 'No tiene Internet']
plt.xlabel('Soporte Internet')
plt.bar(categorias, clientes_TechSupport, color='skyblue')
plt.show()
```

```python
# Clientes que abandonaron segun el medio de pago

servicios = clientes_sin_faltante['PaymentMethod'].unique().astype(str)

clientes_PaymentMethod = []

for x in servicios:
  query = 'SELECT * FROM clientes WHERE Churn = "Yes" AND PaymentMethod = '+'"'+x+'"'
  abandonaron = pd.read_sql(sql=text(query), con=engine.connect())
  clientes_PaymentMethod.append(len(abandonaron))

plt.ylabel('Cantidad de clientes que abandonaron')
plt.title('Clientes que abandonaron segun el tipo de Medio de Pago')
categorias = ['Chek Mail', 'Chequeo\nElectronico', 'Tarjeta\nde Credito','Transferencia']
plt.xlabel('Medio de Pago')
plt.bar(categorias, clientes_PaymentMethod, color='skyblue')
plt.show()
```
